```python
total_by_year = boston_data.sort_values(by='opening')[['buildstart','opening','closure','length']]
```

```python
total_by_year.reset_index(inplace=True)
```

```python
total_by_year.drop('index',inplace=True,axis=1)
```

```python
total_by_year.head()
```

```python
years_df = pd.DataFrame(np.arange(1834,2030,1))
```

```python
years_df.rename(columns={0:'Year'})
```

```python
years_totals = pd.concat([years_df,total_by_year],ignore_index=True, axis=1)
```

```python
years_totals.rename(columns={0:'Year',1:'buildstart',2:'opening',3:'closure',4:'length'},inplace=True)
years_totals.head()
```

```python
years_totals['opening'].values==(years_totals['Year'])
```

```python
years_totals[years_totals['Year']<=1973]['Total_milage_change'].sum()
```

```python
def sum_of_milage(x):
    return years_totals[years_totals['opening'] == x]['length'].sum() - years_totals[years_totals['closure'] == x]['length'].sum()

def total_length(x):
    return years_totals[years_totals['Year']<=x]['Total_milage_change'].sum()
    
```

```python
years_totals['Total_milage_change'] = years_totals['Year'].apply(lambda x: sum_of_milage(x))
```

```python
years_totals['Total_milage'] = years_totals['Year'].apply(lambda x: total_length(x))
```

```python
years_totals.head()
```

```python
fig5, ax5 = plt.subplots(figsize=(18,10))

sns.lineplot(data=years_totals,x = 'Year',y='Total_milage_change')
```

```python
fig6, ax6 = plt.subplots(figsize=(18,10))

sns.lineplot(data=years_totals,x = 'Year',y='Total_milage')
```
